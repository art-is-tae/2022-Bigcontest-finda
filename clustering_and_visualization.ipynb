```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import *
import itertools
import gc

pd.options.display.max_columns = None
```

```python
features_for_clustering = ['loan_rate', 'credit_score', 'desired_amount', 'birth_year', 'loan_limit', 'yearly_income']
```

```python
df = pd.read_csv('../input/big-contest/data_0.csv')
df[features_for_clustering].describe().apply(lambda s: s.apply('{0:.2f}'.format))
```

```python
def cluster(data, n_clusters, clustering_feature, n_features):
    feature_names = ', '.join(clustering_feature)
    
    X = data[clustering_feature]
    
    print(f'=========={feature_names}==========')
    
    ms = KMeans(n_clusters=n_clusters, random_state=42)
    clustered = ms.fit_predict(X)
    
    data['cluster'] = clustered
    print(data['cluster'].value_counts())
```

```python
clustering_features = list(itertools.combinations(features_for_clustering, 3))
clustering_features = [list(t) for t in clustering_features]
clustering_features.sort()
print(len(clustering_features))
```

```python
plt_df = df.sample(100000).copy()
plt_df_idx = plt_df.index
```

```python
for cf in clustering_features:
    cluster(df, 3, cf, 3)
    
    fig, axes = plt.subplots(3, 3, figsize=(12, 12))
    
    plt_df = df.iloc[plt_df_idx].copy()

    for i, f_x in enumerate(cf):
        for j, f_y in enumerate(cf):
            if i == j:
                continue
            else:
                axes[i][j].scatter(plt_df[f_x], plt_df[f_y], c=plt_df['cluster'])
                axes[i][j].set_xlabel(f_x)
                axes[i][j].set_ylabel(f_y)
    plt.show()
```
