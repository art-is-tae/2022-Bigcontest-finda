```python
def change_dtype(df):
    for c in df.columns:
        if 'int64' == str(df[c].dtype):
            df[c] = df[c].astype(np.int32)
        elif 'float64' == str(df[c].dtype):
            df[c] = df[c].astype(np.float32)
```
